## 1. Logistic Classification

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, accuracy_score, classification_report

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, train_size=0.80 ,random_state =30)


# fit a logistic regression model and store the class predictions
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(solver='newton-cholesky', max_iter=1000)

logreg.fit(X_train, y_train.values.ravel())

predictions = logreg.predict(X_test)

# print('Score of logistic regression classifier on test set: {:.2f}\n'.format(logreg.score(X_test, y_test)))
print('accuracy_score pf logistic regression classifier on test set is :{:.2f}\n'.format(accuracy_score(y_test,predictions)))

ConfusionMatrixDisplay.from_estimator(logreg, X_test, y_test)
plt.title('ConfusionMatrix of Logistic')
plt.show()
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions, target_names=logreg.classes_))
```

## 4. SVM

```python
from sklearn.svm import SVC
svc_model=SVC(random_state=0, tol=1e-5)
svc_model.fit(X_train,y_train)

scv_pred=svc_model.predict(X_test)

print('Score : {:.2f}\n'.format(svc_model.score(X_test, y_test)))
print('accuracy_score  :{:.2f}\n'.format(accuracy_score(y_test,scv_pred)))

ConfusionMatrixDisplay.from_estimator(svc_model, X_test, y_test)
plt.title('ConfusionMatrix of SVC')
plt.savefig('confusion_SVC.jpg')
plt.show()

print(classification_report(y_test, scv_pred, target_names=svc_model.classes_))

```
